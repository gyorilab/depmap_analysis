```python
#base_url = 'https://127.0.0.1:5000/'
base_url = 'https://network.indra.bio/'
bfs_url = base_url + 'bfs_search'
reg_url = base_url + 'multi_interactors'
```

```python
query = {'source': 'CTSL', 'reverse': True, 'max_results': 100, 'depth_limit': 4,
         'terminal_ns': ['CHEBI'],
         'node_filter': ['HGNC', 'PUBCHEM', 'CHEBI'],
         'node_blacklist': ['Nucleus Pulposus'],
         'max_per_node': 3,
         'sign': 1}
res = requests.post(bfs_url, json=query)
print(res.status_code)
```

```python
from indra_db import get_primary_db, get_ro
from indra_db.client.readonly.query import HasAgent, FromMeshId
from indra.statements import stmts_from_json
ro = get_ro('primary')
```

```python
# Query the INDRA Database for statements involving CTSL and from papers tagged with the MESH ID
# for "Coronavirus", D017934 (see https://www.ncbi.nlm.nih.gov/mesh/68017934)
q = HasAgent('CTSL') & FromMeshId('D017934')
stmt_res = q.get_statements(ro, limit=10)
stmt_res.statements()
```

```python
# Drill down into a statement about how CTSL activates SARS virus
ctsl_sars = stmt_res.statements()[2]
ctsl_sars
```

```python
# Evidence sentences describing how CTSL activates SARS viral entry
for e in ctsl_sars.evidence:
    print(e.text, '\n')
```
